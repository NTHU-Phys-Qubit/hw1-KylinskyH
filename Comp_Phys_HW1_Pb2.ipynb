```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random as rd
```

```python
#First kind 
def dx():
    return np.random.random()-0.5
def dy():
    return np.random.random()-0.5

def single_walk(step):
    x=[] 
    y=[]
    for i in range(step):
        x+=[dx()]
        y+=[dy()]
    for i in range(step-1):
        x[i+1]= x[i]+x[i+1]
        y[i+1]= y[i]+y[i+1]
    return x,y

def distribution(ensemble,step):
    end_x=[]
    end_y=[]
    for i in range(ensemble):
        x,y= single_walk(step)
        end_x+=[x[-1]]
        end_y+=[y[-1]]
    return end_x,end_y
def final_r(end_x,end_y):
    r=[]
    for i in range(len(end_x)):
        r+=[end_x[i]**2+end_y[i]**2]
    plt.hist(r,bins='auto')
    return 

plt.figure(figsize = (15,4))
plt.subplot(1,3,1)
x,y= single_walk(1000)
plt.plot(x,y)
plt.title("Single Rnadom Walk")

plt.subplot(1,3,2)
end_x,end_y= distribution(2000,500)
color=(0,0,0)
plt.scatter(end_x,end_y,s=0.5)
plt.title("Endpoints distribuiton")

plt.subplot(1,3,3)
final_r(end_x,end_y)
plt.title("Final r")
```
