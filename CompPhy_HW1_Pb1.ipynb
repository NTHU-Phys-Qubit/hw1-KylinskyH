```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import random as rd
```

```python
def dx(): #the displacement of size in the range [-0.5,0.5)
    return rd.random()-0.5

def walk(n): #return a list of x-coord at each step
    x=[0 for i in range(n)]
    for i in range(n-1):
        x[i+1]=x[i]+dx()
    return x
    
def single_rd_walk(n): #plot a single random walk of n steps in step-x diagram
    x= walk(n)
    y=[i for i in range(n)]
    p1= plt.plot(x,y)
    return p1

def hist(ensemble,step): #Draw the histogram of the endpoints of an ensemble of walks
    endpoint=[]
    for i in range(ensemble):
        endpoint+= [walk(step)[step-1]]#the last element of the list walk
    p= plt.hist(endpoint,bins='auto')          #Namely, the endpoint of a walk                         
    return p


plt.figure(figsize = (20,7)) #The first argument 20 specifies the horizontal size.
plt.subplot(1,2,1)
ax= single_rd_walk(1000)
plt.title("One Single Random Walk of 1000 steps")

plt.subplot(1,2,2)
ax2= hist(5000,500)
plt.title("The Histogram of Endpoints of an Esemble of Rand. Walks")
plt.plot()
```

```python
#Second-type Random Walk
# dx=+1 with probability 0.5 and dx=-1 with probability 0.5
def dx():
    if(rd.random()-0.5 >=0):
        return 1
    return -1

def single_rd_walk(n):
    x=[0 for i in range(n)]
    for i in range(n-1):
        x[i+1]=x[i]+dx()
    return x

def print_walk(n):
    x= single_rd_walk(n)
    y=[i for i in range(n)]
    return plt.plot(x,y)

def distribution(ensemble,step):
    endpoints=[]
    for i in range(ensemble):
        endpoints+= [single_rd_walk(step)[-1]]
    p= plt.hist(endpoints,bins='auto')
    return p



plt.figure(figsize= (20,7))
plt.subplot(1,2,1)
print_walk(1000)
plt.title("Single Random Walk of 1000 steps")

plt.subplot(1,2,2)
distribution(5000,500)
plt.title("Endpoints Histogram")
plt.show()
```
